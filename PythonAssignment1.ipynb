### ii) Numeric Data Types and Conversion Functions

#### A Python program that takes an input number from the user, converts it to different numeric data types (integer, float, and complex), and displays the converted values and Explaining the difference between these data types in comments.

```python
# Python program to convert input number to different numeric data types

# Take input from the user
user_input = input("Enter a number: ")

try:
    # Convert the input to integer
    int_value = int(user_input)
    print(f"Integer value: {int_value}")

    # Convert the input to float
    float_value = float(user_input)
    print(f"Float value: {float_value}")

    # Convert the input to complex (use 0 as the imaginary part if not provided)
    complex_value = complex(user_input)
    print(f"Complex value: {complex_value}")

except ValueError:
    print("Invalid input! Please enter a valid number.")

#Differences between numeric data types:

#1. Integer (int):
#   - Represents whole numbers without any fractional or decimal part.
#   - Example: 5, -3, 0
#   - Operations with integers are precise with no decimal point involved.

#2. Float (float):
#   - Represents numbers that have a fractional part (i.e., decimal points).
#   - Example: 5.0, -3.14, 0.0
#   - Used for calculations that require precision with decimal places.

#3. Complex (complex):
#   - Represents numbers in the form of 'a + bj', where 'a' is the real part and 'b' is the imaginary part.
#   - Example: 2 + 3j, -1 - 4j
#   - Used in mathematical computations involving complex numbers where 'j' represents the imaginary unit.

```

### iv) Using the format Method

```python
# Python script to calculate the average of three numbers

# Function to calculate the average of three numbers
def calculate_average(num1, num2, num3):
    return (num1 + num2 + num3) / 3

# user input to enter three numbers
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    num3 = float(input("Enter the third number: "))

# Calculate the average
    average = calculate_average(num1, num2, num3)

# Display the result using % formatting
    print("The average of the three numbers is: %.2f" % average)

except ValueError:
    print("Invalid input! Please enter numerical values.")
```

#### a) A Python program that asks the user for a number and determines whether it is positive, negative, or zero.

```python
# Python program to determine if a number is positive, negative, or zero

# prompt the user to enter a number
try:
    number = float(input("Enter a number: "))

# Determine if the number is positive, negative, or zero
    if number > 0:
        print("The number is positive.")
    elif number < 0:
        print("The number is negative.")
    else:
        print("The number is zero.")

except ValueError:
    print("Invalid input! Please enter a valid numerical value.")
```

 #### b) Implementing a loop that continues to ask the user for a number until they enter 'exit'. Using break to exit the loop and continue to prompt for a new number if the input is not 'exit'.

```python
# Program to determine if a number is positive, negative, or zero
# and to continue prompting until "exit" is entered

def main():
    while True:
# Prompt the user for input
        user_input = input("Enter a number (or type 'exit' to quit): ")
        
# Check if the user wants to exit
        if user_input.lower() == 'exit':
            print("Exiting the program.")
            break
        
# Attempt to convert the input to a number
        try:
            number = float(user_input)
        except ValueError:
            print("Invalid input. Please enter a valid number or type 'exit' to quit.")
            continue
        
# Determine if the number is positive, negative, or zero
        if number > 0:
            print("The number is positive.")
        elif number < 0:
            print("The number is negative.")
        else:
            print("The number is zero.")

if __name__ == "__main__":
    main()
```

#### ii) Relational and Logical Operators

#### A Python script that takes two numbers as input and prints whether both numbers are even, odd, or one of each using relational and logical operators.

```python
def check_even_odd(number):
    return number % 2 == 0

def main():
# Take input from the user
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

# Check if the numbers are even or odd
    num1_even = check_even_odd(num1)
    num2_even = check_even_odd(num2)

# Determine the relationship between the two numbers
    if num1_even and num2_even:
        print("Both numbers are even.")
    elif not num1_even and not num2_even:
        print("Both numbers are odd.")
    else:
        print("One number is even and one number is odd.")

if __name__ == "__main__":
    main()
```

#### iii) For Loop and Bitwise Operators

#### A Python program that takes an integer input and prints its binary, octal, and hexadecimal equivalents using a for loop and bitwise operators.

```python
def int_to_binary(n):
    if n == 0:
        return "0"
    binary_rep = ""
    while n > 0:
        binary_rep = str(n & 1) + binary_rep
        n >>= 1
    return binary_rep

def int_to_octal(n):
    if n == 0:
        return "0"
    octal_rep = ""
    while n > 0:
        octal_rep = str(n & 7) + octal_rep
        n >>= 3
    return octal_rep

def int_to_hexadecimal(n):
    if n == 0:
        return "0"
    hex_chars = "0123456789ABCDEF"
    hex_rep = ""
    while n > 0:
        hex_rep = hex_chars[n & 15] + hex_rep
        n >>= 4
    return hex_rep

def main():
    # Take integer input from the user
    num = int(input("Enter an integer: "))

    # Convert the integer to binary, octal, and hexadecimal
    binary_rep = int_to_binary(num)
    octal_rep = int_to_octal(num)
    hexadecimal_rep = int_to_hexadecimal(num)

    # Print the results
    print(f"Binary: {binary_rep}")
    print(f"Octal: {octal_rep}")
    print(f"Hexadecimal: {hexadecimal_rep}")

if __name__ == "__main__":
    main()
```
